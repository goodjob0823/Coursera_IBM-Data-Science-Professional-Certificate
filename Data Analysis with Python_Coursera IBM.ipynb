```python
# Pipelines (the process to get a prediction)(Normalization -> Polynomial Transform -> Linear Regression)(前2個屬於 transformation，最後一個為 prediction)
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

Input = [('scale',StandardScaler()),('polynomial',PolynomialFeatures(degree = 2)),('mode',LinearRegression())]
pipe = Pipeline(Input)

pipe.train(df_new[['curb-weight','engine-size','highway-mpg']], df_new['price'])
yhat = pipe.predict(df_new[['curb-weight','engine-size','highway-mpg']])

#AttributeError: 'Pipeline' object has no attribute 'train'(無法使用？？？)
```
